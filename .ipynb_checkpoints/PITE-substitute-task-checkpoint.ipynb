```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

boston = load_boston()

df = pd.DataFrame(boston.data, columns=boston.feature_names)
print(df.head())

```

```python
df.hist(column=['RM','PTRATIO'], bins=50)
```

```python
X = df
Y = boston.target
regr = LinearRegression()
regr.fit(X,Y)
plt.scatter(df['RM'], Y)
plt.xlabel('RM')
plt.ylabel('Price')
plt.show()

Y_predict = regr.predict(X)
plt.scatter(df['RM'], Y_predict)
plt.xlabel('RM')
plt.ylabel('Predicted prices')
plt.show()
```

```python
plt.scatter(df['PTRATIO'], Y)
plt.xlabel('PTRATIO')
plt.ylabel('Price')
plt.show()

Y_predict = regr.predict(X)
plt.scatter(df['PTRATIO'], Y_predict)
plt.xlabel('PTRATIO')
plt.ylabel('Predicted prices')
plt.show()
```

In the case of a column 'RM' there is visible linear relationship between its values and target values. The plot shows a positive coorelation - target value increases with the increase of RM value. 
When it comes to column 'PTRATIO' we don't see any strongly marked linear relationship with target values. 

```python
mseFull = np.mean((Y-regr.predict(X))**2)
print(mseFull)
```
